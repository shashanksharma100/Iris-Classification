**Importing Libraries**

```python
import cv2
from google.colab.patches import cv2_imshow
```

**Orignal Picture**

```python
image = cv2.imread("golden.png.jpg")
cv2_imshow(image)
```

**Convert the RBG image to grayscale**

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

**Invert the grayscale image**

```python
invert_img = cv2.bitwise_not(gray_img)
cv2_imshow(invert_img)
```

**Apply a Guassian Blur**

```python
smoothing_img = cv2.GaussianBlur(invert_img, (21,21), sigmaX=0, sigmaY=0)
cv2_imshow(smoothing_img)
```

**Blend the grayscale image with blurred negative**

```python
def dodgeV2(x, y):
 return cv2.divide(x, 255-y, scale=256)
```

```python
final_img = dodgeV2(gray_img, smoothing_img)
cv2_imshow(final_img)
```
